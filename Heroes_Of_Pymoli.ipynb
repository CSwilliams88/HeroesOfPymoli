```python
import pandas as pd
import numpy as np
```

```python
file="purchase_data.csv"
```

```python
#Create my main Dataframe
file_df = pd.read_csv(file)
file_df.head()
```

```python
index = file_df.index
totalrows = len(index)
print (totalrows)
```

```python
#Calculate number of players
total_players = len(file_df["SN"].value_counts())
total_players
```

```python
#Display total number of players
number_of_players = pd.DataFrame({"Total Players":[total_players]})
number_of_players
```

```python
#Calculating total unique items, average price, total purchases, total revenue
total_uitems = len(file_df["Item ID"].unique())
total_uitems

avg_price = round(file_df["Price"].mean(),2)
avg_price

total_purchases = (file_df["Purchase ID"].count())
total_purchases

total_revenue = (file_df["Price"].sum())
total_revenue
```

```python
#Displaying that summary
summary_df = pd.DataFrame({"Number of Unique Items": [total_uitems], "Average Price": [avg_price], "Number of Purchases": [total_purchases], "Total Revenue": [total_revenue]})
summary_df
```

```python
#Taking the SN, Age, Gender columns from the main data set and making sure no duplicates
gender_df = file_df[["SN", "Age", "Gender"]].drop_duplicates()
gender_df
```

```python
#Getting a count for each gender
gender_count = gender_df["Gender"].value_counts()
gender_count
```

```python
gender_percent = gender_count / total_players * 100
```

```python
gender_summary = pd.DataFrame({"Total Players":gender_count, "Percentage":gender_percent})
```

```python
#Displaying Summary
gender_summary
```

```python
#Formatting the percentage column to show as a percentage 
gender_summary.style.format({"Percentage":"{:.2f}%"})
```

```python
#Grouping the data set by Gender
gender_group = file_df.groupby("Gender")
gender_group.head()
```

```python
#Acquiring the purchase count, avg purchase price, total purchase price, and total average 
purchase_count = gender_group["Purchase ID"].count()
purchase_count
```

```python
avg_purchase = gender_group["Price"].mean()
avg_purchase
```

```python
total_purchase = gender_group["Price"].sum()
total_purchase
```

```python
total_average = total_purchase / gender_count
total_average
```

```python
#Displaying data
purchasing_analysis = pd.DataFrame({"Purchase Count":purchase_count, 
                                    "Average Purchase Price($)":avg_purchase,
                                    "Total Purchase Price($)":total_purchase,
                                    "Total Average Purchase per Person($)": total_average})
purchasing_analysis
```

```python
#Creating my bins and bin labels
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
#Adding the Age Demo column and defining what it is
file_df["Age Demographics"] = pd.cut(file_df["Age"], bins, labels=group_labels)

file_df
```

```python
#Grouping the data by Age demo
group_df = file_df.groupby("Age Demographics")
group_df.head()
```

```python
#Getting a count for each group but making sure a SN only gets counted once
group_count = group_df["SN"].nunique()
group_count.head()
```

```python
group_percentage = group_count / total_players * 100
group_percentage.head()
```

```python
#Displaying the count and percentage but making sure the percentage column displays as a percentage
age_demo = pd.DataFrame({"Total Count":group_count,
                         "Percentage of Players":group_percentage})
age_demo.style.format({"Percentage of Players":"{:.2f}%"})
```

```python
#Calculating the purchase count per age demo
age_purchase_count = group_df["Purchase ID"].count()
age_purchase_count
```

```python
#gathering the average, total, and total average for each age demographic
age_average_purchase = group_df["Price"].mean()
age_average_purchase
```

```python
age_total_purchase = group_df["Price"].sum()
age_total_purchase
```

```python
age_total_average = age_total_purchase / group_count
age_total_average
```

```python
#Displaying that data set
age_purchase_analysis = pd.DataFrame({"Purchase Count":age_purchase_count, 
                                    "Average Purchase Price($)":age_average_purchase,
                                    "Total Purchase Price($)":age_total_purchase,
                                    "Total Average Purchase per Person($)":age_total_average})
age_purchase_analysis
```

```python
#Grouping the main data set by the SN and then getting the purchase count, average, and top purchase value
top_spenders = file_df.groupby("SN")
top_spenders.head()
```

```python
top_count = top_spenders["SN"].count()
top_count
```

```python
top_avg_purchase_price = top_spenders["Price"].mean()
top_avg_purchase_price
```

```python
top_purchase_value = top_spenders["Price"].sum()
top_purchase_value
```

```python
#Displaying the data set but sorting the columns in  a descending order
top_spender = pd.DataFrame({"Purchase Count":top_count,
                            "Average Purchase Price":top_avg_purchase_price,
                            "Total Purchase Value":top_purchase_value})
decending_top_spender = top_spender.sort_values(["Total Purchase Value"], ascending=False).head()
decending_top_spender.head()

```

```python
#Grouping the main data set by item ID
popular_group = file_df.groupby("Item ID")
popular_group.head()
```

```python
#Finding the purchase count for each item
popular_count = popular_group["Item ID"].count()
popular_count
```

```python
#Getting the total sum for each item
popular_purchase = popular_group["Price"].sum()
popular_purchase
```

```python
item_name = popular_group["Item Name"]
item_price= popular_group["Price"]
```

```python
#displaying data set in descending order by the purchase count
popular_groups = pd.DataFrame({"Item Name":item_name,
                            "Purchase Count":popular_count,
                            "Price":item_price,
                            "Total Value":popular_purchase})
decending_popular_groups = popular_groups.sort_values(["Purchase Count"], ascending=False).head()
decending_popular_groups
```

```python
#Displaying the data set in descending order by the total value
popular_groups = pd.DataFrame({"Item Name":item_name,
                            "Purchase Count":popular_count,
                            "Price":item_price,
                            "Total Value":popular_purchase})
decending_popular_groups = popular_groups.sort_values(["Total Value"], ascending=False).head()
decending_popular_groups
```
